### Basic image plotting

```python
# Import necessary modules

import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Black and White Image
img_array = np.zeros(shape = (6,8,1), dtype = "uint8")
img_array[3,4] = 128  # Set to a mid-range value for gray
plt.imshow(img_array, cmap='gray', vmin=0, vmax=255)
plt.title("Random black and white image")
plt.axis(False)
plt.show()
```

```python
# Colour Image
img_array = np.zeros(shape = (6,8,3), dtype="uint8")
img_array[4,3] = (54,124,82)
plt.imshow(img_array)
plt.title("Random colour image")
plt.axis(False)
plt.show()
```

```python
try:
    image = cv2.imread("Images/boeing.jpeg")
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image, vmin=0, vmax=255)
    plt.axis(False)
    plt.title("Boeing 747 - Queen of the Skies")
    plt.show()
except:
    print("Image not found")
```

```python
try:
    image = cv2.imread("Images/boeing.jpeg")
    image = image.astype(np.int16)
    image = image - 42
    image = np.clip(image, 0, 255)
    image = image.astype(np.uint8)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image, vmin = 0, vmax = 255)
    plt.axis(False)
    plt.title("Boeing 747 - Queen of the Skies")
    plt.show()
except:
    print("Image not found")
```
